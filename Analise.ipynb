# Extraindo Informações - Candidatos

Por conta da falta de informações extraíveis dos candidatos, acabamos optando por extraír:

1. **Quantidade de Instâncias**
2. **Candidatos por Estado**:
    - Quantidade de candidatos para governador por estado
    - Média da quantidade de candidatos
    - Desvio Padrão da quantidade de candidatos
3. **Candidatos por Partido**:
    - Quantidade de candidatos para governador por partido
    - Média da quantidade de partidos
    - Desvio Padrão da quantidade de partidos

```python
df = pd.read_csv('Resultados\\Candidatos.csv', delimiter=',', encoding='utf-8')
df.head()
```

### Quantidade de Instâncias

```python
total = df['SQ_CANDIDATO'].count()
print('Quantidade total de candidatos:', total)
```

### Quantidade por Estado

```python
uf = df['SG_UF'].value_counts()

uf_counts = uf.reset_index()
uf_counts.columns = ['SG_UF', 'COUNT']
```

```python
uf_mean = uf.mean()
uf_std = uf.std()
```

```python
print(f"Média da quantidade de candidatos por estado: {uf_mean:.2f}")
print(f"Desvio padrão da quantidade de candidatos por estado: {uf_std:.2f}")
print('Quantidade total de candidatos por estado:')
uf_counts
```

### Quantidade por Partido

```python
partido = df['SG_PARTIDO'].value_counts()

partido_counts = uf.reset_index()
partido_counts.columns = ['SG_PARTIDO', 'COUNT']
```

```python
partido_mean = partido.mean()
partido_std = partido.std()
```

```python
print(f"Média da quantidade de partidos por estado: {partido_mean:.2f}")
print(f"Desvio padrão da quantidade de partidos por estado: {partido_std:.2f}")
print('Quantidade total de candidatos por partido:')
partido_counts
```
